### Load Libraries

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.utils.data as data
```

### Data processing

```python
# load ascii text and covert to lowercase
filename = 'wonderland.txt'
raw_text = open(filename, 'r', encoding='utf-8').read()
raw_text = raw_text.lower()

# create mapping of unique char to int
chars = sorted(list(set(raw_text)))
char_to_int = dict((c, i) for i, c in enumerate(chars))

# summarize the loaded data
n_chars = len(raw_text)
n_vocab = len(chars)
print("Total Characters: ", n_chars)
print("Total Vocab: ", n_vocab)

# Prepare the datasets of input to output pairs encoded as int
seq_length = 100
dataX = []
dataY = []
for i in range(0, n_chars - seq_length, 1):
  seq_in = raw_text[i:i + seq_length]
  seq_out = raw_text[i + seq_length]
  dataX.append([char_to_int[char] for char in seq_in])
  dataY.append(char_to_int[seq_out])

n_patterns = len(dataX)
print("Total patterns: ", n_patterns)

# Reshape X to be [sample, time steps, features]
X = torch.tensor(dataX, dtype=torch.float32).view(n_patterns, seq_length, 1)
X = X / float(n_vocab)
y = torch.tensor(dataY)
```

```python
class CharModel(nn.Module):
  def __init__(self):
    super().__init__()
    self.lstm = nn.LSTM(input_size=1, hidden_size=256, num_layers=2, batch_first=True)
    self.dropout = nn.Dropout(0.2)
    self.linear = nn.Linear(256, n_vocab)

  def forward(self, x):
    x, _ = self.lstm(x)
    # take only the last output of the sequence
    x = x[:, -1, :]
    # produce output
    x = self.linear(self.dropout(x))
    return x
```

### Test

```python
n_epoch = 40
batch_size = 128
model = CharModel()

optimizer = optim.Adam(model.parameters())
loss_fn = nn.CrossEntropyLoss(reduction='sum')
loader = data.DataLoader(
    data.TensorDataset(X, y),
    shuffle=True,
    batch_size=batch_size,
    pin_memory=torch.cuda.is_available(),
)

best_model = None
best_loss = np.inf
for epoch in range(n_epoch):
  model.train()
  for X_batch, y_batch in loader:
    X_batch = X_batch.to(device, non_blocking=True)
    y_batch = y_batch.to(device, non_blocking=True)
    y_pred = model(X_batch)
    loss = loss_fn(y_pred, y_batch)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
  # Validation
  model.eval()
  loss = 0
  with torch.no_grad():
    for X_batch, y_batch in loader:
      X_batch, y_batch = X_batch.to(device), y_batch.to(device)
      y_pred = model(X_batch)
      loss += loss_fn(y_pred, y_batch)
    if loss < best_loss:
      best_loss = loss
      best_model = model.state_dict()
    print("Epoch %d: Cross-entropy: %.4f" % (epoch, loss))

torch.save([best_model, char_to_int], "single-char.pth")
```
